```python
import csv
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path
```

Define the file 

```python
filename = "data/fires 2021/modis_2021_Australia.csv"
```

Checking the file headers

```python
# with open(filename) as f:
#     reader = csv.reader(f)
#     header_row = next(reader)
    
# Since using with requires to access the open file inside this function, instead path will be used  
path = Path("data/fires 2021/modis_2021_Australia.csv")
lines = path.read_text().splitlines()
reader = csv.reader(lines)
header_row = next(reader)

header_row
```

Extract the brightness from the file

```python
bright = []
brightness_row = header_row.index('brightness')
for row in reader:
    bright.append(row[brightness_row])
    
bright
```

Ploting the data on the map

```python
fig = go.Figure(data=go.Scatter(x=header_row, y=bright))

# plot the data
fig.show()
```
